```python
import numpy as np
import pandas as pd
import os
import sys
```

```python
WHITE_list = np.array([14, 15, 20, 21, 23, 29, 31, 32, 34, 36, 38, 40, 42, 45, 47, 48, 64, 69, 79, 80, 82, 85, 86, 91, 92, 95, 97])
ASIAN_list = np.array([5, 7, 13, 24, 25, 27, 35, 37, 43, 44, 46, 49, 51, 52, 54, 55, 56, 57, 61, 63, 65, 68, 71, 72, 73, 74, 75, 76, 81, 84, 93, 94])
URG_list = np.array([22, 70, 78, 28, 60, 58])


print('WHITE_list size: {}'.format(len(WHITE_list)))
print('ASIAN_list size: {}'.format(len(ASIAN_list)))
print('URG_list size: {}'.format(len(URG_list)))
```

### RandomForest

### random_partition1

```python
root_path = 'YOUR_PATH/NuripsDataSet2021/experiments/FixedTrainValSplit_generic_models/AsianSubset/RandomForest/binary/'
```

```python
scenario_21vs6_summary_path = '21vs6/seed2_summary'

```

```python
scenario_21vs4_df = pd.read_csv(os.path.join(root_path, scenario_21vs6_summary_path, 'AllSubjects_summary.csv'))
```

```python
scenario_21vs4_WHITE_df = scenario_21vs4_df.loc[scenario_21vs4_df['subject_id'].isin(WHITE_list)]
scenario_21vs4_ASIAN_df = scenario_21vs4_df.loc[scenario_21vs4_df['subject_id'].isin(ASIAN_list)]
scenario_21vs4_URG_df = scenario_21vs4_df.loc[scenario_21vs4_df['subject_id'].isin(URG_list)]
```

```python
scenario_21vs4_WHITE_df
```

```python
scenario_21vs4_ASIAN_df
```

```python
scenario_21vs4_URG_df
```

```python
print('RF, Binary, Average Accuracy:')
print('WHITE: {}'.format(np.mean(scenario_21vs4_WHITE_df.corresponding_test_accuracy.values)))
print('ASIAN: {}'.format(np.mean(scenario_21vs4_ASIAN_df.corresponding_test_accuracy.values)))
print('URG: {}'.format(np.mean(scenario_21vs4_URG_df.corresponding_test_accuracy.values)))
```
